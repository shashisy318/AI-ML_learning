# Working on AQI Dataset

```python
# AQI dataset
data = pd.read_csv("globalAirQuality.csv")
data

data.head()
data.tail()

data.describe()
```

```python
# selecting data in given dataset

data["city"]    # for selecting a column data
data[["city","aqi"]]   # for selecting more than one column data

data.loc[2]  # for selecting row through label
data.loc[0:2]  # for selecting multiple rows through label

data.iloc[2]  # for selecting row through index 
data.iloc[0:2]  # for selecting multiple rows through index (end exclusive)

# cells - row,column
data.loc[0:2,["city","aqi"]]  # for selecting row and column

data.iloc[0:3,1:5]   
# data.iloc[0:3,["city","aqi","country"]]  # will show error (reason below)
#iloc accepts only numerical value as the indices are numerical hence need to put columns in form of thier index not value

# cell - selecting single cell (scalar)
data.at[0, "city"]   # select single cell through label
data.iat[0, 2]     # # select single cell through index
```

```python
# filtering of data 

data[data["aqi"] > 100]

data[(data["aqi"] > 100) & (data["temperature"] > 30)]

aqi_data = data[(data["aqi"] > 100) & (data["temperature"] > 30)][["city","aqi"]]

# aqi_data.loc[0]   # will show error cause there is no any label with '0' as pandas preserve the label of dataset
aqi_data.iloc[0]

aqi_data.loc[6]   #label '6' is present after the filter hence will not show error


# filtering of data by querrying - return copy

data.query("aqi > 100 & temperature >30")[["city","aqi"]]

aqi_val = 50
data.query("aqi > @aqi_val")
```

```python
# cleaning the data(handle missing values)

df = pd.read_csv("raw_data.csv")

# isnull() / isna()
df.isnull()   # returns true if it's null value
# df.isna()  # same as isnull()

# isnull().sum()
df.isnull().sum()  # counts no of null values in every column

# dropna()
df.dropna()   # drops row having null value
df.dropna(axis=1) # drops column having null value

# fillna()
df.fillna(0)     # fill passed data at the place of null values

cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean()

cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)
cleaned_data

# ffill() / bfill()
df.ffill()   # it fill the last encountered value at the place of null value from forward
df.bfill()  # it fill the last encountered value at the place of null value from backward
```

```python
# cleaning the data(handle duplicate values)

df = pd.read_csv("raw_data.csv")

# duplicated()
df.duplicated()   # for overall data

df[["country","gender"]].duplicated()  # for specific column

# drop_duplicates()
df2 = df.copy()   # store copy so that will be able to recover 

df.drop_duplicates()   # it does not make changes in original dataset
df.drop_duplicates(inplace = True)  # it make changes in original dataset

df = df2 # for gaining datset back
```

```python
# cleaning the data(handle datatypes & date_time)

# dtypes
df.dtypes   #returns datatype of each column

#astype()
df2 = df.copy()
df2 = df2.fillna(0)
df2["age"] = df2["age"].astype("int64")
print(df2["age"].dtypes)


# to_datetime  -> converts into datetime(timestamp)

date_str = "2025-12-31"
new_dtstr = pd.Series(pd.to_datetime(date_str))
new_dtstr
```

```python
# cleaning the data(handle strings)

# str.lower() / str.upper() / str.capitalize()
df["gender"].str.capitalize()

# str.strip()
df["name"].str.strip()

# str.split()
df["name"].str.split("@")

# str.contains("xyz")

df["country"].str.contains("india", case = False)
```
